```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings_function = OllamaEmbeddings(
    model="nomic-embed-text"
)

query_test = "What is the largest planet"

query_embeddings = embeddings_function.embed_query(query_test)
```

```python
print(query_embeddings)
```

```python
docs = [
    "The sun is the center of our solar system.",
    "Mercury is the smallest planet in our solar system.",
    "Venus is similar in size to Earth but very hot.",
    "Mars is known as the red planet because of iron oxide.",
    "Jupiter is the largest planet and has a Great Red Spot.",
]

embedding_functions = OllamaEmbeddings(model="nomic-embed-text")
```

```python
embedding_functions
```

```python
doc_embed = embedding_functions.embed_documents(docs)
```

```python
len(doc_embed)
```

Vector Database

```python
from langchain_chroma import Chroma
```

```python
vector_store = Chroma.from_texts(
    texts = docs,
    embedding = embedding_functions
)
```

```python
print(vector_store.get(include=["embeddings","documents"]))
```

```python
query = "Which planet  is the biggest?"
results = vector_store.similarity_search(query)
```

```python
print(results)
```

```python
print(results[0].page_content)
```

```python
def generate_response(query:str, context_docs:str) -> str:
    context = context_docs

    #create the prompt using a template
    prompt_template = f"""
    You are a helpful assistant for a space encyclopedia.
    Answer the following question based only on the context provided below.
    If the information is not in the context, say "I don't have that information in my knowledge base."

    CONTEXT:
    {context}

    QUESTION:
    {query}

    ANSWER:
    """

    resp = ollama.chat(
        model='llama3.2:1b',
        messages=[
            {
                "role":"user",
                "content":prompt_template
            }
        ]
    )

    return resp['message']['content']
```

```python
query= "What is at the centre of our solar system?"
results = vector_store.similarity_search(query)

final_answer = generate_response(query,results[0].page_content)
```

```python
print(final_answer)
```
